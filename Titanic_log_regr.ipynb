```python
Cab = X['Cabin']
```

```python
Cab 
```

```python
sectors = []
for i in range(len(Cab)):
    if ord(str(Cab[i])[0]) == 110:
        sectors.append(78-65-6)
    else:
        sectors.append(ord(str(Cab[i])[0])-65)
sectors
```

```python
side = []
for i in range(len(Cab)):
    if str(Cab[i])[-1] == 'P':
        side.append(1)
    else:
        side.append(0)
side
```

```python
sleep_place = pd.DataFrame(train_data['CryoSleep']).values.ravel()
sleep_place_correct = []
for i in range(len(sleep_place)):
    if str((sleep_place[i]))=='True':
        sleep_place_correct.append(1)
    elif str(sleep_place[i])=='False':
        sleep_place_correct.append(-1)
    else:
        sleep_place_correct.append(0)
sleep_place_correct
```

```python
#X_new =
print(type(sleep_place_correct))
print(type(side))
print(type(sectors))
```

```python
X_new = [sleep_place_correct, side, sectors]
columns_new = ['sleep_place_correct', 'side', 'sectors']
X_new = pd.DataFrame(X_new).T
X_new =  X_new.rename(columns={0 : 'sleep_place_correct', 1 : 'side', 2 : 'sectors'})
```
