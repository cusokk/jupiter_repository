```python
#'test_tit.csv'  'train_tit.csv'
import pandas as pd
train_data = pd.read_csv('train_tit.csv')
```

```python
train_data
```

```python
CryoSleep_train = pd.DataFrame(train_data['CryoSleep']).values.ravel()
```

```python
CryoSleep_train_cycle = []
for i in range(len(CryoSleep_train)):
    if str(CryoSleep_train[i])=='True':
        CryoSleep_train_cycle.append(1)
    elif str(CryoSleep_train[i])=='False':
        CryoSleep_train_cycle.append(0)
    else:
        if i%2==0:
            CryoSleep_train_cycle.append(1)
        else:
            CryoSleep_train_cycle.append(0)
```

```python
CryoSleep_train_cycle
```

```python
Cabin_train = pd.DataFrame(train_data['Cabin']).values.ravel()
Corpus_train = []
```

```python

for i in range(len(Cabin_train)):
    if ord(str(Cabin_train[i])[0]) == 110:
        Corpus_train.append(78-65-6)
    else:
        Corpus_train.append(ord(str(Cabin_train[i])[0])-65)
```

```python
Corpus_train_norm = []
for i in range(len(Corpus_train)):
    norm_object_i = (Corpus_train[i] - min(Corpus_train))/(max(Corpus_train)-min(Corpus_train))
    Corpus_train_norm.append(norm_object_i)
Corpus_train_norm
```

```python
b = 'd/43/s'

def number_between(a):
    a = a[2:]
    a = (a[-3::-1])[::-1]
    return(a)
print(number_between(b))
```

```python
room_number_train = []
for i in range(len(Cabin_train)):
    if pd.isnull(Cabin_train[i]):
        room_number_train.append(0)
    else:
        current_between = number_between(Cabin_train[i])
        room_number_train.append(int(current_between))
room_number_train
```

```python
room_number_train_norm = []
for i in range(len(room_number_train)):
    current_number = (room_number_train[i]-min(room_number_train))/(max(room_number_train)-min(room_number_train))
    room_number_train_norm.append(current_number)
room_number_train_norm
```

```python
side_train = []
for i in range(len(Cabin_train)):
    if pd.isnull(Cabin_train[i]):
        side_train.append(i%2) 
    elif Cabin_train[i][-1] == 'P':
        side_train.append(1)
    else:
        side_train.append(0)
side_train
```

```python
y = (pd.DataFrame(train_data['Transported']).values.ravel()).astype('int')
```

```python
y
```

```python
'''
side_train
room_number_train_norm
Corpus_train_norm
CryoSleep_train_cycle
'''
X_new_train = [CryoSleep_train_cycle, Corpus_train_norm, room_number_train_norm, side_train]
X_new_train = pd.DataFrame(X_new_train).T
X_new_train =  X_new_train.rename(columns={0 : 'CryoSleep', 1 : 'Corpus', 2 : 'room' , 3 : 'side'})
X_new_train 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(n_neighbors=2000, p=2)
neigh.fit(X_new_train, y)
```
